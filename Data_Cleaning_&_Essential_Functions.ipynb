3. Explain the use of text functions such as TRIM , LEFT, RIGHT, MID, and CONCAT in data cleaning.

->Use of Text Functions in Data Cleaning
Introduction

Data cleaning is an essential step in data analysis, especially when working with textual data. Text functions in tools like MS Excel help in removing errors, extracting useful information, and organizing data into a consistent format. Functions such as TRIM, LEFT, RIGHT, MID, and CONCAT are commonly used for this purpose.
Text Functions and Their Uses in Data Cleaning
1. TRIM Function
Purpose: Removes unnecessary spaces from text
Use in Data Cleaning:
Eliminates extra spaces at the beginning, end, and between words
Ensures consistency in text fields like names and addresses
Example:
Input: " Sonu Kumar "
Output: "Sonu Kumar"
2. LEFT Function
Purpose: Extracts characters from the beginning of a text string
Use in Data Cleaning:
Used to extract prefixes such as department codes or initials
Example:
Input: "IT2024"
Formula: LEFT(A1,2)
Output: "IT"
3. RIGHT Function
Purpose: Extracts characters from the end of a text string
Use in Data Cleaning:
Helps extract suffixes like ID numbers or last digits
Example:
Input: "EMP5678"
Formula: RIGHT(A1,4)
Output: "5678"
4. MID Function
Purpose: Extracts characters from the middle of a text
Use in Data Cleaning:
Useful for extracting specific parts such as dates, codes, or embedded values
Example:
Input: "EMP1234"
Formula: MID(A1,4,2)
Output: "12"
5. CONCAT Function (or CONCATENATE)
Purpose: Combines two or more text strings into one
Use in Data Cleaning:
Merges data from multiple columns (e.g., first name and last name)
Helps create complete and meaningful fields
Example:
Input: "Sonu" and "Kumar"
Formula: CONCAT(A1," ",B1)
Output: "Sonu Kumar"
Conclusion

Text functions like TRIM, LEFT, RIGHT, MID, and CONCAT play a vital role in data cleaning. They help in removing inconsistencies, extracting relevant information, and organizing text data efficiently. Proper use of these functions ensures that the dataset is clean, structured, and ready for accurate analysis.

 ->Understanding TODAY() and NOW() in Excel
a. Difference between TODAY() and NOW()
.TODAY()
Returns only the current date
Does not include time
Example Use:
Tracking attendance date
Generating daily reports
.NOW()
Returns current date and time
Updates automatically
Example Use:
Recording login time
Tracking transaction timestamps
b. Days Left Until a Specific Date
Given: A1 = 2025-06-10
Formula:
=A1 - TODAY()
Explanation:
Calculates the number of days remaining from today to the given date.
c. Display Current Date and Time in Format
Formula:
=TEXT(NOW(),"MM/DD/YYYY HH:MM AM/PM")
Output Example:
03/28/2026 10:45 AM
d. What Happens When Worksheet is Reopened?
If a cell contains =TODAY():
The date will automatically update to the current date
Example:
Today: 28/03/2026
Tomorrow: 29/03/2026 (updates automatically)
e. Store Static Date (Non-changing Date)
Keyboard Shortcut:
Ctrl + ; (semicolon)
Explanation:
Inserts the current date as a fixed value
Does not change like TODAY()
Conclusion
TODAY() → current date
NOW() → current date + time
Both are dynamic functions used for real-time data tracking.
